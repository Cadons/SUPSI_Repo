## Introduzione

Scarica il dataset `Fremont_Bridge_Bicycle_Counter.csv` dal link sotto, e importalo come segue

```python
import pandas as pd
df = pd.read_csv('data/Fremont_Bridge_Bicycle_Counter.csv')
df["DateT"] = pd.to_datetime(df["Date"])
df = df.sort_values("DateT").groupby("DateT").first()
df = df[["Fremont Bridge East Sidewalk","Fremont Bridge West Sidewalk"]].rename(columns={
    "Fremont Bridge East Sidewalk":"East",
    "Fremont Bridge West Sidewalk":"West",
})
df
```

Qui trovi ulteriori informazioni sul dataset.

Risposta:ricampiona i dati per ogni giorno sommando i relativi passaggi a east e ovest

```python
dfd = df.resample(rule="D").sum()
import plotly.express as px
px.line(dfd, x=dfd.index, y="East")
```

## Esercizio

Elabora i dati del dataset, eventualmente generando delle visualizzazioni, per rispondere alle seguenti domande.
- Come impatta il giorno della settimana sul transito di biciclette?
- Quale e' stato l'impatto della pandemia nel transito di bici?
- Come e' cambiato l'uso della bici negli anni?
- Come cambia l'uso della bici nelle varie stagioni?

```python
# Soluzione
days = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

dw=df.groupby([pd.to_datetime(df.index).day_name()]).sum().reindex(days)
fig=px.histogram(dw,x=dw.index,y=[dw["East"],dw["West"]],barmode="group")
fig.layout["xaxis"]["title"]="Day"
fig.layout["yaxis"]["title"]="Number of bikes"
fig.layout["legend"]["title"]["text"]="Legend"

fig.show()
```

```python
dp=df
dp=dp.resample("M").sum()
px.line(dp)
```

```python
seasons=["Winter","Spring","Summer","Autumn"]


dw=df.resample("3M").sum()
dw["Season"]=seasons[0]
dw.loc[(dw.index.month >= 3) & (dw.index.month <= 5),"Season"]=seasons[1]
dw.loc[(dw.index.month >= 6)&(dw.index.month <= 8),"Season"]=seasons[2]
dw.loc[(dw.index.month >= 9)&(dw.index.month <= 11),"Season"]=seasons[3]
```

```python

fig=px.histogram(dw,x="Season",y=[dw["East"],dw["West"]],barmode="group")
fig.layout["xaxis"]["title"]="Season"
fig.layout["yaxis"]["title"]="Number of bikes"
fig.layout["legend"]["title"]["text"]="Legend"
fig.show()
```
